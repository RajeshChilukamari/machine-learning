```python
import random,math
import numpy as np
import matplotlib.pyplot as plt
```

```python
##1. Initialization

w=np.array([0.2,0.9,0.4,0.7,0.6,0.5,0.8,0.3]).reshape(4,2)
#w=np.random.rand(4,2)
print(w)
lr=0.5
plt.plot(w)
```

```python
##2. Distance Calculation

def distcal(w,x):
    d=np.zeros((2,), dtype='float')
    for i in range(2):
        for j in range(len(w)):
            #print(pow(w[j][i]-x[j],2))
            d[i] += pow(w[j][i]-x[j],2)
    return d
```

```python
##3. Finding minimum distance index

def mindist(d):
    for i in range(len(d)-1):
        if (d[i] < d[i+1]):
            j=i
        else:
            j=i+1
    return j
    
```

```python
##4. weight updation

def wtupdt(w,x,j):
    for i in range(len(w)):
        w[i][j] = w[i][j]+lr*(x[i]-w[i][j])
    return w
```

```python
for q in range(1):
    x = np.array([0,0,1,1])
    #x = np.random.randint(0,2,(1,4))

    print("\nx: {}".format(x))

    plt.plot(x,w)

    ##2. Distance Calculation
    d = distcal(w,x)
    print("\nDistance calculation:\n {}".format(d))


    ##3. Finding minimum distance index
    j = mindist(d)
    print("\nMinimum Distance Index: {}".format(j))


    ##4. weight updation
    w = wtupdt(w,x,j)
    print("\nnew weight updation:\n {}".format(w))
```
